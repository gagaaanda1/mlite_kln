# Konversi SQLite ke MySQL
Notebook ini membantu mengonversi dump SQL SQLite menjadi sintaks yang lebih kompatibel dengan MySQL.


```python
from pathlib import Path
import re
import shutil

src = Path(r'f:\KLINIK\BECKUP\main.sql')
out = Path(r'c:\laragon\www\mlite_kln\main.mysql.sql')
bak = src.with_suffix('.sqlite.sql')

if not bak.exists():
    shutil.copy2(src, bak)

text = src.read_text(encoding='utf-8-sig')
text = re.sub(r'(?im)^PRAGMA\s+foreign_keys\s*=\s*false;\s*\n?', '', text)
text = text.replace('INTEGER PRIMARY KEY AUTOINCREMENT', 'INT AUTO_INCREMENT PRIMARY KEY')
text = text.replace('INTEGER PRIMARY KEY', 'INT PRIMARY KEY')
text = text.replace('AUTOINCREMENT', 'AUTO_INCREMENT')
text = text.replace('"', '`')
text = text.replace('`;', '`')
text = text.replace('BEGIN TRANSACTION;', '')
text = text.replace('COMMIT;', '')
out.write_text(text, encoding='utf-8')
print(f'Backup: {bak}')
print(f'Output: {out}')
```
